```python
import numpy as np
import pandas as pd
```

```python
df1=pd.DataFrame(np.random.randint(1,33,size=(10000,5)), columns=["s1","s2","s3","s4","s5"])
df2=pd.DataFrame(np.random.randint(34,40,size=(10000,5)), columns=["s1","s2","s3","s4","s5"])
df3=pd.DataFrame(np.random.randint(41,50,size=(10000,5)), columns=["s1","s2","s3","s4","s5"])
df4=pd.DataFrame(np.random.randint(51,60,size=(10000,5)), columns=["s1","s2","s3","s4","s5"])
df5=pd.DataFrame(np.random.randint(61,70,size=(10000,5)), columns=["s1","s2","s3","s4","s5"])
df6=pd.DataFrame(np.random.randint(71,80,size=(10000,5)), columns=["s1","s2","s3","s4","s5"])
df7=pd.DataFrame(np.random.randint(81,100,size=(10000,5)), columns=["s1","s2","s3","s4","s5"])
df=pd.concat([df1,df2,df3,df4,df5,df6,df7])
df.head()
```

```python
print(df1.info())
df1.head()
```

```python
df["Obtain"]=df.loc[:,"s1":"s5"].apply(sum,axis=1)
```

```python
df
```

```python
df["Total"]=500
df.head()
```

```python
df["Persentage"]= df.Obtain / df.Total * 100
```

```python
df.head()
```

```python
def grade(x):
    if x>=80:
        return "A+"
    elif x>=70:
        return "A"
    elif x>=60:
        return "B"
    elif x>=50:
        return "C"
    elif x>=40:
        return "D"
    elif x>=33:
        return "E"
    else:
        return "FAIL"
df["Grade"]=df.Persentage.apply(grade)    
    
```

```python
df.head()
```

```python
df.corr().plot(kind="bar")
```

```python
dic={
    "FAIL":7,
    "E":6,
    "D":5,
    "C":4,
    "B":3,
    "A":2,
    "A+":1
}
df["Grade1"]=df.Grade.map(dic)
```

```python
df.head()
```

```python
df[["Persentage","Grade1"]].corr().plot(kind="bar")
```

```python
df.corr().plot(kind="bar")
```

```python
df1=df[["s1","s2","s3","s4","s5","Grade"]]
df1.head(2)
df1=df1.sample(frac=1)
df1
```

```python
x=df1.loc[:,"s1":"s5"]
x.head()
```

```python
y=df1["Grade"]
```

```python
y.head()
```

```python
y=pd.get_dummies(y)
y.head()
```

```python
y.tail()
```

```python
y.drop_duplicates()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam,SGD
```

```python
model=Sequential()
```

```python
model.add(Dense(32,activation="relu", input_shape=(5,)))
model.add(Dense(16,activation="relu", input_shape=(5,)))
model.add(Dense(7,activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer=SGD(lr=0.002),
              loss="categorical_crossentropy",
              metrics=["acc"])
```

```python
model.fit(x,y, epochs=30)
```
